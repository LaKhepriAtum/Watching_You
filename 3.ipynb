```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Watching_You\\sigudong.csv')
df
```

```python
file_list = df['법정동명'].values.tolist()
file_list
```

```python
sido = []
gu =[]
dong = []


for file in file_list:
    if   10 < len(file)  :
         a,b,c = file.split(' ')
         sido.append(a)
         gu.append(b)
         dong.append(c)
    else :
        pass

df['sido'] = sido
df['gu'] = gu
df['dong'] = dong

df.to_csv('./sigu.csv',encoding='utf-8-sig')
```
